# 数据分布可视化

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python

fig=plt.figure(figsize=(13,8))
fig.set(alpha=0.2)
plt.subplot2grid((2,3),(0,0))             # 在一张大图里分列几个小图
titanic_data.Survived.value_counts().plot(kind='bar')# 柱状图 
plt.title("Survived") # 标题
plt.ylabel("Amount")  

plt.subplot2grid((2,3),(0,1))
titanic_data.Pclass.value_counts().plot(kind="bar")
plt.ylabel("Amount")
plt.title("Pclass")

plt.subplot2grid((2,3),(0,2))
plt.scatter(titanic_data.Survived, titanic_data.Age)
plt.ylabel("Age")                         # 设定纵坐标名称
plt.grid(b=True, which='major', axis='y') 
plt.title("Age (1 means Survived)")


plt.subplot2grid((2,3),(1,0), colspan=2)
titanic_data.Age[titanic_data.Pclass == 1].plot(kind='kde')   
titanic_data.Age[titanic_data.Pclass == 2].plot(kind='kde')
titanic_data.Age[titanic_data.Pclass == 3].plot(kind='kde')
plt.xlabel("Age")# plots an axis lable
plt.ylabel("Density") 
plt.title("Age distribution of passengers of all levels")
plt.legend(('First class', 'Second class','Three class'),loc='best') # sets our legend for our graph.


plt.subplot2grid((2,3),(1,2))
titanic_data.Embarked.value_counts().plot(kind='bar')
plt.title("Amount of embarked")
plt.ylabel("Amount")  
plt.show()
```

# 特征与结果之间的关系

## 与船舱等级关系

```python
fig=plt.figure()
fig.set(alpha=0.2) #set alpha make pic looks better（change line appha）
dead=titanic_data.Pclass[titanic_data.Survived == 0].value_counts() #this is series 
survived=titanic_data.Pclass[titanic_data.Survived == 1].value_counts()
df=pd.DataFrame({'survived':survived,'dead':dead}) #create dataframe
df
```

```python
#use padas dataframe.plot() method that wrapper arround matplotlib.plot()
df.plot(kind='bar')
plt.xlabel("ticket class")
plt.ylabel("peopel")
plt.title("Relationship between surviving number and ticket class")
plt.show()
#we can see pclass==1 survival rate better
```

## 与性别关系

```python
fig=plt.figure()
fig.set(alpha=0.1) #set alpha make pic looks better（change line appha）
men=titanic_data.Survived[titanic_data.Sex == 'male'].value_counts() #this is series 
women=titanic_data.Survived[titanic_data.Sex == 'female'].value_counts()#value_counts() return unique values
#men
df=pd.DataFrame({'men':men,'women':women}) #create dataframe
df.plot.bar(stacked='true')#Histogram stacked default false
plt.xlabel("sex type")
plt.ylabel("amount")
plt.title("Relationship between surviving number and type of sex")
plt.show()
#most of the surivors are women(lady and kid first)
```

```python
def trans_to_num(df):
    print('hello')
    dummies_Cabin=pd.get_dummies(df["Cabin"],prefix="Cabin")
    dummies_Embarked = pd.get_dummies(df["Embarked"], prefix="Embarked")
    dummies_Sex = pd.get_dummies(df["Sex"], prefix="Sex")
    dummies_Pclass = pd.get_dummies(df["Pclass"], prefix="Pclass")
#     #cat new series
    cat_df=pd.concat([df,dummies_Cabin,dummies_Embarked,dummies_Sex,dummies_Pclass],axis=1)
    result_df=cat_df.drop(['Pclass', 'Name', 'Sex', 'Ticket', 'Cabin', 'Embarked'], axis=1)
    return result_df
train_data=trans_to_num(titanic_data)
train_data
```
